```python
import os

# model = VGG19(weights='imagenet', include_top=False)

def get_img_files(base_path):
    file_list = []

    folder_path = f'data/paraiba/{base_path}'


    for file_name in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, file_name)):
            file_list.append(file_name)
    return (base_path, sorted(file_list))
```

```python
langs_to_check = ['pt', 'en', 'es', 'de', 'it', 'ru', 'zh', 'fr']


img_lang_files = map(get_img_files, langs_to_check)
```

```python
import os

diretorio_principal = "data/paraiba"

# Coletando todos os caminhos de arquivo em uma lista
caminhos_arquivos = []

for subpasta, _, arquivos in os.walk(diretorio_principal):
    for arquivo in arquivos:
        caminho_arquivo = os.path.join(subpasta, arquivo)
        caminhos_arquivos.append(caminho_arquivo)

# Gerando a matriz de combinações
combinacoes = []

for caminho1 in caminhos_arquivos:
    for caminho2 in caminhos_arquivos:
        if caminho1 != caminho2:  # Evitando combinar um arquivo com ele mesmo
            combinacoes.append([caminho1, caminho2])

# Agora a variável "combinacoes" contém todas as combinações possíveis
print(combinacoes)
```
